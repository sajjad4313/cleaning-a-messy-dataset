```python
df['VOTES'].replace(',','', regex=True, inplace=True)
df['VOTES']=df['VOTES'].astype(str).astype(float)
df.info()
```

```python
df['VOTES'].fillna(df['VOTES'].mean() ,inplace=True)
df.isnull().sum()
```

```python
df['RATING'].fillna(df['RATING'].mean() ,inplace=True)
df.isnull().sum()
```

```python
df.dropna(inplace=True)
df.isnull().sum()
```
